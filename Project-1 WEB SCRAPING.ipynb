```python
abc="https://www.flipkart.com/search?q=ac+1+ton&sid=j9e%2Cabm%2Cc54&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=RECENT&suggestionId=ac+1+ton%7CAir+Conditioners&requestId=6008eb92-0bed-4dcf-bc09-630415aee7aa&as-searchtext=ac"
prices=[]
a=[]
brands=[]
models=[]
room_size=[]
power=[]
star=[]
for i in range(1,23):
    url=abc+str(i)
    print(url)
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    p=soup.find_all("div",class_="_30jeq3 _1_WHN1")
    
    info=soup.find_all("div",class_="_4rR01T")
    for i in info:
        a.append(i.text)
    
    
    #brands
    for i in a:
        brands.append(i.split()[0])
    
    #models
    for i in a:
        models.append(i.split("Model")[0])
        
    #prices
    for i in p:
        prices.append(i.text)
        
 
    #star
    for i in a:
        if "1 Ton" in i:
            star.append(i.split("1 Ton")[1].split(" Star ")[0])
    
    #power
    
    for i in data:
        a=re.findall(r"\d+.\sW|\d+.\.\sW|\d+.\.\d+\skWh",i.text)
        if a:
            #print(a)
            power.append(a)
    
    
    
    #room_size
    for i in data:
        a=re.findall(r"\d+.\ssqft",i.text)
        if len(a)>0:
            #print(a)
            room_size.append(a)
    
    
    
    print(len(brands))
    print(len(models))
    print(len(prices))
    print(len(star))
    print(len(power))
    print(len(room_size))
```

```python
AC={"brands":brands,
   "models":models,
   "prices":prices,
   "star":star,
   "power":power,
   "room_size":room_size}
```
